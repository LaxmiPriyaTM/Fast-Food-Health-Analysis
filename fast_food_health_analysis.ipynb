## 📌 Introduction

In today's fast-paced world, fast food consumption has increased significantly. This project analyzes how dietary habits and lifestyle factors affect health indicators such as BMI, energy levels, and overall health score.

The goal is to use data science techniques to identify patterns and relationships between these variables.

## 🎯 Objective

- Analyze the impact of fast food consumption on BMI  
- Study the relationship between lifestyle factors and health  
- Visualize patterns using data analysis techniques  

## 📂 Dataset Description

The dataset contains the following features:

- Age  
- Gender  
- Fast Food Meals Per Week  
- Average Daily Calories  
- BMI  
- Physical Activity Hours Per Week  
- Sleep Hours Per Day  
- Energy Level Score  
- Digestive Issues  
- Doctor Visits Per Year  
- Overall Health Score  

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 📥Load Dataset

```python
data = pd.read_csv("fast_food_consumption_health_impact_dataset.csv")
data.head()
```

## 🧹 Data Cleaning

The dataset is checked for missing values and duplicate entries to ensure data quality.

```python
print(data.isnull().sum())  
print(data.duplicated().sum())  
print(data.columns)
```

## 🔍 Exploratory Data Analysis

```python
data.describe()
```

## 📈 Fast Food vs BMI

This section analyzes the relationship between fast food consumption and BMI.

```python
plt.scatter(data['Fast_Food_Meals_Per_Week'], data['BMI'])
plt.xlabel("Fast Food Meals Per Week")
plt.ylabel("BMI")
plt.title("Fast Food vs BMI")
plt.show()
```

### 💡 Insight
Higher fast food consumption is associated with increased BMI levels.

## 📈 Calories vs BMI

```python
plt.scatter(data['Average_Daily_Calories'], data['BMI'])
plt.xlabel("Daily Calories")
plt.ylabel("BMI")
plt.title("Calories vs BMI")
plt.show()
```

### 💡 Insight
Higher calorie intake contributes to higher BMI values.

## 📈 Physical Activity vs Health Score

```python
plt.scatter(data['Physical_Activity_Hours_Per_Week'], data['Overall_Health_Score'])
plt.xlabel("Physical Activity Hours")
plt.ylabel("Health Score")
plt.title("Activity vs Health")
plt.show()
```

### 💡 Insight
Increased physical activity leads to better health scores.

## 📈 Sleep vs Energy Level

```python
plt.scatter(data['Sleep_Hours_Per_Day'], data['Energy_Level_Score'])
plt.xlabel("Sleep Hours")
plt.ylabel("Energy Level")
plt.title("Sleep vs Energy")
plt.show()
```

### 💡 Insight
Adequate sleep improves energy levels and daily performance.

## 🔥 Correlation Matrix

```python
import seaborn as sns

numeric_data = data.select_dtypes(include=['number'])

sns.heatmap(numeric_data.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

### 💡 Insight
The heatmap shows relationships between variables, highlighting how lifestyle factors influence health indicators.

## 📌 Key Insights

- Fast food consumption increases BMI  
- Higher calorie intake leads to weight gain  
- Physical activity improves health  
- Sleep enhances energy levels  

## ✅ Conclusion

This study demonstrates that lifestyle factors such as diet, physical activity, and sleep significantly impact health. Data-driven analysis helps in understanding these relationships and promoting healthier living habits.
